Supongamos que somos un agente de Airbnb localizado en Lisboa, y tenemos que atender peticiones de varios clientes. Tenemos un archivo llamado airbnb.csv (en la carpeta data) donde tenemos informaci√≥n de todos los alojamientos de Airbnb en Lisboa.

```python
airbnb = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/REPASO 2022/data/airbnb.csv")
airbnb.head()
```

En concreto el dataset tiene las siguientes variables:

* room_id: el identificador de la propiedad 
* host_id: el identificador del due√±o de la propiedad
* room_type: tipo de propiedad (vivienda completa/(habitacion para compartir/habitaci√≥n privada)
* neighborhood: el barrio de Lisboa
* reviews: El numero de opiniones
* overall_satisfaction: Puntuacion media del apartamento
* accommodates: El numero de personas que se pueden alojar en la propiedad
* bedrooms: El n√∫mero de habitaciones
* price: El precio (en euros) por noche

**Caso 1**

Alicia va a ir a Lisboa durante una semana con su marido y sus 2 hijos. Est√°n buscando un apartamento con habitaciones separadas para los padres y los hijos. No les importa donde alojarse o el precio, simplemente quieren tener una experiencia agradable. Esto significa que solo aceptan lugares con m√°s de 10 cr√≠ticas con una puntuaci√≥n mayor de 4. Cuando seleccionemos habitaciones para Alicia, tenemos que asegurarnos de ordenar las habitaciones de mejor a peor puntuaci√≥n. Para aquellas habitaciones que tienen la misma puntuaci√≥n, debemos mostrar antes aquellas con m√°s cr√≠ticas. Debemos darle 3 alternativas.

```python
alicia = airbnb[    
  # lugares con m√°s de 10 cr√≠ticas
  (airbnb.reviews > 10) &

  #puntuaci√≥n mayor de 4
  (airbnb.overall_satisfaction > 4) &

  # apartamento con habitaciones separadas
  (airbnb.bedrooms >= 2)

# ordenar las habitaciones de mejor a peor puntuaci√≥n
].sort_values(
    by=["overall_satisfaction","reviews"], 
    ascending=False
    
# Debemos darle 3 alternativas.
).head(3)

alicia
```
